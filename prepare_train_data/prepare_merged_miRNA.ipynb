```python
import pandas as pd
import numpy as np
import gc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_mir_tcga = pd.read_parquet('/mnt/jack-5/amismailov/miRNA_study/PROPER_NORMALIZED_DATASETS/TCGA_miR_CPM_expanded.parquet')
df_mir_gtex = pd.read_parquet('/mnt/jack-5/amismailov/miRNA_study/PROPER_NORMALIZED_DATASETS/GTEx_adaptive_K/GTEx_CPM_boot.parquet')
```

```python
df_mir_gtex = df_mir_gtex.T
df_mir_gtex.head()
```

```python
df_mir_tcga.head()
```

```python
lower_ = lambda x: x.lower()
df_mir_tcga.index = list(map(lower_, df_mir_tcga.index.tolist()))
df_mir_gtex.index = list(map(lower_, df_mir_gtex.index.tolist()))
```

```python
gtex_names = df_mir_gtex.index.tolist()
tcga_names = df_mir_tcga.index.tolist()
```

```python
print(f'GTEx miRs len: {len(gtex_names)}')
print(f'TCGA miRs len: {len(tcga_names)}')
print(f'Intersection: {len(set(gtex_names) & set(tcga_names))}')
```

```python
intersect = set(gtex_names) & set(tcga_names)
intersect = sorted(list(intersect))
```

```python
len(intersect)
```

```python
df_mir_tcga = df_mir_tcga.loc[intersect]
df_mir_gtex = df_mir_gtex.loc[intersect]
```

```python
len(df_mir_tcga)
```

```python
assert all(df_mir_tcga.index == df_mir_gtex.index)
```

```python
pd_mir_merged = pd.concat([df_mir_tcga, df_mir_gtex], axis=1)
pd_mir_merged.shape
```

```python
pd_mir_merged.sum(axis=0)
```

```python
pd_mir_merged.to_parquet('/mnt/jack-5/amismailov/miRNA_study/PROPER_NORMALIZED_DATASETS/merged/merged_expanded_miR_v2.parquet')
```
